```python
import pandas as pd
```

```python
def kualitas(df) : 
  servis = []
  for i in range(0,len(df),1) : 
    x = df['servis'][i]
    if x <= 5 : 
      servis.append([[1,"jelek"]])
    elif x > 5 and x < 10 :
      a = -1*(x-10)/(10-5)
      b = (x-5)/(10-5)
      servis.append([[a,"jelek"],[b,"cukup"]])
    elif x >= 10 and x <= 30 : 
      servis.append([[1,"cukup"]])
    elif x > 30 and x < 50 : 
      a = -1*(x-50)/(50-30)
      b = (x-30)/(50-30)
      servis.append([[a,"cukup"],[b,"bagus"]])
    elif x >= 50 and x <= 70 :
      servis.append([[1,"bagus"]])
    elif x > 70 and x < 90 : 
      a = -1*(x-90)/(90-70)
      b = (x-70)/(90-70)
      servis.append([[a,"bagus"],[b,"sangat bagus"]])
    elif x>=90 : 
      servis.append([[1,"sangat bagus"]])

  return servis
```

```python
def harga(df) : 
  price = []
  for i in range(len(df)) : 
    x = df['harga'][i]
    if x <= 2 :
      price.append([[1,"murah"]])
    elif x > 2 and x < 4 : 
      a = -1*(x-4)/(4-2)
      b = (x-2)/(4-2)
      price.append([[a,"murah"],[b,"sedang"]])
    elif x >= 4 and x <= 6 : 
      price.append([[1,"sedang"]])
    elif x > 6 and x < 9 : 
      a = -1*(x-9)/(9-6)
      b = (x-6)/(9-6)
      price.append([[a,"sedang"],[b,"mahal"]])
    elif x >= 9 :
      price.append([[1,"mahal"]])

  return price
```

```python
def fuzzy(servis,harga) : 
  fuz = []
  for i in range(len(servis)) :
    temp = [] 
    for j in range(len(servis[i])) : 
      for k in range(len(harga[i])) : 
        if servis[i][j][1] == "jelek" and harga[i][k][1] == "murah" : 
          if servis[i][j][0]<=harga[i][k][0] : 
            temp.append([servis[i][j][0],"rendah"])
          else :
            temp.append([harga[i][j][0],"rendah"])
        elif servis[i][j][1] == "jelek" and harga[i][k][1] == "sedang" : 
          if servis[i][j][0]<=harga[i][k][0] : 
            temp.append([servis[i][j][0],"rendah"])
          else :
            temp.append([harga[i][j][0],"rendah"])
        elif servis[i][j][1] == "jelek" and harga[i][k][1] == "mahal" : 
          if servis[i][j][0]<=harga[i][k][0] : 
            temp.append([servis[i][j][0],"rendah"])
          else :
            temp.append([harga[i][j][0],"rendah"])
        elif servis[i][j][1] == "cukup" and harga[i][k][1] == "murah" : 
          if servis[i][j][0]<=harga[i][k][0] : 
            temp.append([servis[i][j][0],"sedang"])
          else :
            temp.append([harga[i][j][0],"sedang"])
        elif servis[i][j][1] == "cukup" and harga[i][k][1] == "sedang" : 
          if servis[i][j][0]<=harga[i][k][0] : 
            temp.append([servis[i][j][0],"sedang"])
          else :
            temp.append([harga[i][j][0],"sedang"])
        elif servis[i][j][1] == "cukup" and harga[i][k][1] == "mahal" : 
          if servis[i][j][0]<=harga[i][k][0] : 
            temp.append([servis[i][j][0],"rendah"])
          else :
            temp.append([harga[i][j][0],"rendah"])
        elif servis[i][j][1] == "bagus" and harga[i][k][1] == "murah" : 
          if servis[i][j][0]<=harga[i][k][0] : 
            temp.append([servis[i][j][0],"tinggi"])
          else :
            temp.append([harga[i][j][0],"tinggi"])
        elif servis[i][j][1] == "bagus" and harga[i][k][1] == "sedang" : 
          if servis[i][j][0]<=harga[i][k][0] : 
            temp.append([servis[i][j][0],"tinggi"])
          else :
            temp.append([harga[i][j][0],"tinggi"])
        elif servis[i][j][1] == "bagus" and harga[i][k][1] == "mahal" : 
          if servis[i][j][0]<=harga[i][k][0] : 
            temp.append([servis[i][j][0],"sedang"])
          else :
            temp.append([harga[i][j][0],"sedang"])
        elif servis[i][j][1] == "sangat bagus" and harga[i][k][1] == "murah" : 
          if servis[i][j][0]<=harga[i][k][0] : 
            temp.append([servis[i][j][0],"tinggi"])
          else :
            temp.append([harga[i][j][0],"tinggi"])
        elif servis[i][j][1] == "sangat bagus" and harga[i][k][1] == "sedang" : 
          if servis[i][j][0]<=harga[i][k][0] : 
            temp.append([servis[i][j][0],"tinggi"])
          else :
            temp.append([harga[i][j][0],"tinggi"])
        elif servis[i][j][1] == "sangat bagus" and harga[i][k][1] == "mahal" : 
          if servis[i][j][0]<=harga[i][k][0] : 
            temp.append([servis[i][j][0],"sedang"])
          else :
            temp.append([harga[i][j][0],"sedang"])
    fuz.append(temp)

  i = 0
  while i < len(fuz) : 
    iter = len(fuz[i])
    j = 0
    while j < iter - 1 : 
      iter = len(fuz[i])
      k = j + 1
      while k < iter : 
        if fuz[i][j][1] == fuz[i][k][1] : 
          if fuz[i][j][0] >= fuz[i][k][0] : 
            hapus = fuz[i][k]
            fuz[i].remove(hapus)
          else : 
            hapus = fuz[i][j]
            fuz[i].remove(hapus)

        iter = len(fuz[i])
        k += 1
      j += 1
    i += 1

  return fuz
```

```python
def defuzzy(fuz) : 
  hasil = []
  for i in range(len(fuz)) : 
    jumlah = 0
    bagi = 0
    for j in range(len(fuz[i])) : 
      if fuz[i][j][1] == "rendah" : 
        jumlah += 30*fuz[i][j][0]
        bagi += fuz[i][j][0]
      elif fuz[i][j][1] == "sedang" : 
        jumlah += 60*fuz[i][j][0]
        bagi += fuz[i][j][0]
      elif fuz[i][j][1] == "tinggi" : 
        jumlah += 90*fuz[i][j][0]
        bagi += fuz[i][j][0]
    hasil.append(jumlah/bagi)
  
  return hasil
```

```python
def peringkat(defuz) : 
  rate = []
  iter = len(defuz)
  i = 0
  while i < iter : 
    if defuz[i] == max(defuz) and len(rate) < 10 : 
      rate.append([i+1,defuz[i]])
      defuz.remove(defuz[i])

    iter = len(defuz)
    i += 1
  return rate
```

```python
#Main Program

df = pd.read_excel('https://github.com/mughidaf/Fuzzy-Tupro-AI-2/blob/main/bengkel.xlsx?raw=true')

quality = kualitas(df)
price = harga(df)

fuzzy_set = fuzzy(quality,price)
crisp_set = defuzzy(fuzzy_set)

bengkel_terbaik = peringkat(crisp_set)

output_id = []
output_crisp_value = []

for i in range(len(bengkel_terbaik)) : 
  output_id.append(bengkel_terbaik[i][0])
  output_crisp_value.append(bengkel_terbaik[i][1])

output = pd.DataFrame({'ID Bengkel':output_id, 'Skor Bengkel': output_crisp_value})
output.to_excel("peringkat.xlsx",index = False)
```
